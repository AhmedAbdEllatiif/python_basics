```python
df = pandas.read_csv("supermarkets.csv")
df
```

```python
df["Address"] = df["Address"] + ', ' + df['City'] + ', ' + df['State'] + ', ' + df['Country']
df
```

```python
from geopy.geocoders import ArcGIS
nom = ArcGIS()
# To Apply function on the whole column
df['Coordinates'] = df["Address"].apply(nom.geocode)
df
```

```python
df['Latitude'] = df['Coordinates'].apply(lambda x: x.latitude)
df['Longitude'] = df['Coordinates'].apply(lambda x: x.longitude)
df
# try this the see the type  >>>> type(df.loc[0,'Coordinates']) <<<<
# Coordinates column is a Location (geopy.location.Location) object (Whiel using DataFrame you store object in a column)
# What if x is None (x is the return value of Coordinates) 
# In this case you must check if the x is None 
# Then store None in the column if the x is None
# df['Latitude'] = df['Coordinates'].apply(lambda x: x.latitude if x != None else None)
```
