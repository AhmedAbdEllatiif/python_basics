```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
r = requests.get("https://www.century21.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/")
c = r.content                          
```

```python
soup = bs(c,"html.parser")
all_prices = soup.find_all("a",{"class" : "listing-price"})
prices = []
for price in all_prices:
    text =price.text
    text =text.strip() # to remove the new lines \n
    prices.append(text)
  
print(prices)
```

```python
all = soup.find_all('div',{'class':'property-card-primary-info'})
print(all)
```

```python
for item in all:
    try:
        b = item.find('div',{'class':'property-address-info'}).find("div",{'class':'property-address'}).text
    except:
        b = None
    print(b)
```

```python
# all is a Resultset so we have to loop on it
data = []
for item in all:
        d = {}
        
         ################### Price ########################
        try:
            d['price'] = item.find_all('a',{'class':'listing-price'})[0].text.strip()
        except:
            d['price'] = None
            
        ################### Bed Rooms ########################
        try:
        except:
            d['bed_rooms'] = None
            
       
        ################### Baths ########################
        try:
            d['baths'] = item.find('div',{'class':'property-baths'}).find('strong').text
        except:
            d['baths'] = None
            
        ################### Half_Baths ########################
        try:
            d['half_baths'] = item.find('div',{'class':'property-half-baths'}).find('strong').text
        except:
            d['half_baths'] = None
        
        
        ################### Area ########################
        try:
            d['area'] = item.find('div',{'class':'property-sqft'}).find('strong').text
        except:
            d['area'] = None
            
        ################### Address ########################
        try:
            d['address'] = item.find('div',{'class':'property-address-info'}).find("div",{'class':'property-address'}).text.strip()
        except:
            d['address'] = None
            
            
        ################### City ########################
        try:
            d['city'] = item.find('div',{'class':'property-address-info'}).find("div",{'class':'property-city'}).text.strip()
        except:
            d['city'] = None
            
            
        ################### type ########################
        try:
            d['type'] = item.find('div',{'class':'pdp-listing-type sale'}).text
        except:
            d['type'] = None
            
            
        data.append(d)
        

print(data)
```
