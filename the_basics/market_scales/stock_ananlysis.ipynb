```python
from pandas_datareader import data
import datetime
from bokeh.plotting import figure,show,output_file
from bokeh.embed import components
from bokeh.resources import CDN

start = datetime.datetime(2020,11,1)
end = datetime.datetime.now()
df = data.DataReader(name="GOOG",data_source="yahoo",start=start,end=end)
df
```

```python
def win_los(c,o):
    if c > o:
        value = "win"
    elif c < o: 
        value = "lose"
    else:
        value = "equal"
    return value

df["Status"] = [win_los(c,o) for c,o in zip(df.Close,df.Open)]
df["Middle"] = (df.Open+df.Close) /2
df["Height"] = abs(df.Open-df.Close)

df
```

```python
win = df.index[df.Status == "win"]
lose = df.index[df.Status == "lose"]

middle_win = df.Middle[df.Status == "win"]
middle_lose = df.Middle[df.Status == "lose"]

height_win = df.Height[df.Status == "win"]
height_lose = df.Height[df.Status == "lose"]
'''
for date in win:
    print(date)
print("\n")    
for date in lose:
    print(date)
'''
```

```python
p = figure(x_axis_type="datetime",width = 1000, height = 300,sizing_mode="scale_width")

# Styling
p.title.text = "Candlestick Chart"
p.grid.grid_line_alpha = 0.3

# segment(heightX,heightY,lowX,lowY)
p.segment(df.index,df.High,df.index,df.Low,color="black")

# p.rect(x,y,width,height)
hours_12 = 12 * 60 * 60 * 1000
p.rect(win,middle_win, hours_12,
       height_win,line_color="black",fill_color="#CCFFFF")

p.rect(lose,middle_lose, hours_12,
        height_lose,line_color="black",fill_color="#FF3333")



#output_file('stock_graph.html')
#show(p)

# Live Url 
# components is a tuple 
jscript,div1 = components(p)
cdn_js = CDN.js_files
cdn_css = CDN.css_files

print(cdn_js[0])
```
